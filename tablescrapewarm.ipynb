```python
df = tables[0]
df.head()
```

```python
df.columns
```

```python
df = df.drop(columns=['Location', 'Gust Knots', 'Vis SM', 'Baro inHg' ])
```

```python
df = df.rename(columns={
    'Name': 'Lentoasema',
    'Wind Knots': 'Tuulta',
    'Temp/Dew C': 'Lämpö/Kosteus',
    'METAR': 'Metadata',
})
df.head()
```

```python
df = df.rename(columns={
    'Station': 'Lyhenne'
})
df.head()
```

```python
df = df.drop(columns=['Tuulta' ])
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['Lämpö/Kosteus'] = df['Lämpö/Kosteus'].str.split('/').str.get(0)
```

```python
df = df.rename(columns={
    'Lämpö/Kosteus': 'Lämpötila, C'
})
df.head()
```

```python
df['Lämpötila, C'] = df['Lämpötila, C'].astype(float)
```

```python
df.dtypes
```

```python
df['Metadata'] = df['Metadata'].str.slice(0, 12).str.split('Z').str.get(0) + 'Z'
df.head()
```

```python
df.to_csv("hot.csv", index=False)
```
