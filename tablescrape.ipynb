```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd

# Käynnistä selain (oletetaan, että käytät Chromea)
driver = webdriver.Chrome()

# Lataa sivu
url = "https://www.suomenpankki.fi/fi/Tilastot/rahapolitiikan-valineet/kuviot/rapo-kuviot-fi/ekp_ja_markkinakorot_pv_chrt_fi/"
driver.get(url)

# Odota hetki, jotta sivu ja iframe latautuvat
driver.implicitly_wait(10)

# Vaihda iframeen
iframe = driver.find_element(By.ID, "ReportIframe")
driver.switch_to.frame(iframe)

# Lataa iframen sisältö BeautifulSoupilla
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Etsi taulukot iframen sisältä
tables = soup.find_all("table")

# Jos haluat muuntaa ne DataFrameiksi
for i, table in enumerate(tables):
    df = pd.read_html(str(table))[0]
    print(f"Taulukko {i}:")
    print(df)

# Sulje selain
driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup

# Käynnistä selain
driver = webdriver.Chrome()

# Lataa sivu
url = "https://www.suomenpankki.fi/fi/Tilastot/rahapolitiikan-valineet/kuviot/rapo-kuviot-fi/ekp_ja_markkinakorot_pv_chrt_fi/"
driver.get(url)

# Odota hetki, jotta iframe latautuu
driver.implicitly_wait(10)

# Vaihda iframeen
iframe = driver.find_element(By.ID, "ReportIframe")
driver.switch_to.frame(iframe)

# Lataa iframen sisältö BeautifulSoupilla
soup = BeautifulSoup(driver.page_source, 'html.parser')

# Sulje selain, emme tarvitse sitä enää
driver.quit()

# Etsi data-elementit ja käsittele ne
# Oletetaan, että data on <div>-elementeissä tai vastaavissa

# Tässä esimerkissä käsitellään taulukot:
tables = soup.find_all("table")
for i, table in enumerate(tables):
    rows = table.find_all("tr")
    for row in rows:
        cols = row.find_all(["td", "th"])
        cols = [col.text.strip() for col in cols]
        print(cols)

# Voit muuntaa kerätyn datan taulukoiksi (esim. pandas DataFrame)
```
